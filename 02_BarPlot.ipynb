Bar Plot

```python
import matplotlib.pyplot as plt
x=["Python","c","c++","Java"]
y=[85,70,60,82]
plt.xlabel("Language", fontsize=10)
plt.ylabel("Popularity", fontsize=10)
plt.title("Language Popularity", fontsize=20)
c=["b","g","r","m"]
plt.bar(x,y,width=0.4,color=c,align="center",edgecolor="black",linewidth=4, linestyle=":",alpha=0.5,label="popularity")

plt.legend()
plt.show()
```

Vertical Bar graph

```python
import matplotlib.pyplot as plt
import numpy as np

x=["Python","c","c++","Java"]
y=[85,70,60,82]
z=[35,40,20,42]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Language", fontsize=10)
plt.ylabel("Popularity", fontsize=10)
plt.title("Language Popularity", fontsize=20)

plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity2")

# plt.xticks(p+width/2,x)
plt.xticks(p+width/2,x,rotation=20)

plt.legend()
plt.show()
```

Horizontal Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np

x=["Python","c","c++","Java"]
y=[85,70,60,82]
z=[35,40,20,42]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Language", fontsize=10)
plt.ylabel("Popularity", fontsize=10)
plt.title("Language Popularity", fontsize=20)

plt.barh(p,y,width,color="r",label="popularity In India")
plt.barh(p1,z,width,color="y",label="popularity In USA")

# plt.xticks(p+width/2,x)
plt.xticks(p+width/2,x,rotation=20)

plt.legend()
plt.show()
```
